```python
# GOAL : time trace of Center of Mass distance for Markov States to ensure 
#        we are sampling different bound states

import pandas as pd
import numpy as np
%pylab inline

# prettify
plt.style.use('default')
plt.rcParams['font.size'] = 20
```

```python
# Loading in distance files generated using ptraj

distance_0 = pd.read_table('./distance_dat/centermass0_seg1_500_ACTUAL.dat', sep = '\s+')
distance_1 = pd.read_table('./distance_dat/centermass1_seg1_500_ACTUAL.dat', sep = '\s+')
distance_2 = pd.read_table('./distance_dat/centermass2_seg1_500_ACTUAL.dat', sep = '\s+')

# Sanity Check
display(distance_0.head(), distance_1.head(), distance_2.head())
```

```python
# Plotting center mass distance for each state

distance_0['center_mass'].plot(figsize = (15,5), label = '0')
distance_1['center_mass'].plot(figsize = (15,5), label = '1')
distance_2['center_mass'].plot(figsize = (15,5), label = '2')
pylab.legend(loc='upper right', prop={'size': 20})

# Plotting mean distance values for each state
plt.axhline(y=distance_0.apply(np.mean)[1], color='blue', linestyle='--')
plt.axhline(y=distance_1.apply(np.mean)[1], color='orange', linestyle='--')
plt.axhline(y=distance_2.apply(np.mean)[1], color='green', linestyle='--')

pylab.xlabel('# Frames', )
pylab.ylabel('Distance ($\AA$)')

#savefig('./pictures/DISTANCE_ANALYSIS.png', dpi = 600, bbox_inches = 'tight')
```
